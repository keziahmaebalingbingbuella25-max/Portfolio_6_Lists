PROGRAM: College GWA and Academic Honors Eligibility Checker
Program Description:
This is my personalized application program that takes a traditional grade calculator 
and upgrades it using List Collections.
Instead of dealing with 18 different loose variables floating around in the background, 
or typing out long lines of math and chains of or operators, 
I bundled my grades and units into two clean list structures: grades_list and units_list. 
By keeping my data inside these single, convenient packages, 
I was able to use our lesson's built-in list functions 
like sum() and max() to let Python handle the heavy lifting automatically, 
completely cleaning up my code equations.

```python
# SECTION 1: DATA INPUT AND TYPE CONVERSION

# I'm collecting the user inputs here using basic assignment statements (=).
# Variable names must start with a letter.
# I used float() because grades and unit values at my university have decimals.

name = input("Enter Student Name: ")

s1_g = float(input("Subject 1 Grade: "))
s1_u = float(input("Subject 1 Units: "))
s2_g = float(input("Subject 2 Grade: "))
s2_u = float(input("Subject 2 Units: "))
s3_g = float(input("Subject 3 Grade: "))
s3_u = float(input("Subject 3 Units: "))
s4_g = float(input("Subject 4 Grade: "))
s4_u = float(input("Subject 4 Units: "))
s5_g = float(input("Subject 5 Grade: "))
s5_u = float(input("Subject 5 Units: "))
s6_g = float(input("Subject 6 Grade: "))
s6_u = float(input("Subject 6 Units: "))
s7_g = float(input("Subject 7 Grade: "))
s7_u = float(input("Subject 7 Units: "))
s8_g = float(input("Subject 8 Grade: "))
s8_u = float(input("Subject 8 Units: "))
s9_g = float(input("Subject 9 Grade: "))
s9_u = float(input("Subject 9 Units: "))


# SECTION 2: LIST PACKAGING

# This is where I apply the core lesson, instead of keeping 18 loose variables, 
# I am packaging them into two single, organized array constants.
# I remembered that list constants must use square brackets [ ] and commas.

grades_list = [s1_g, s2_g, s3_g, s4_g, s5_g, s6_g, s7_g, s8_g, s9_g]
units_list = [s1_u, s2_u, s3_u, s4_u, s5_u, s6_u, s7_u, s8_u, s9_u]


# SECTION 3: NUMERIC EXPRESSIONS & INDEX LOOKUP 

# To compute my total points, I am using List Indexing and Lookup.
# I have to remember that Python counts coordinates starting from 0, not 1
# I am matching index positions [0 through 8] to multiply pairs together.
total_points = (grades_list[0] * units_list[0]) + (grades_list[1] * units_list[1])+ \
               (grades_list[2] * units_list[2]) + (grades_list[3] * units_list[3]) + \
               (grades_list[4] * units_list[4]) + (grades_list[5] * units_list[5]) + \
               (grades_list[6] * units_list[6]) + (grades_list[7] * units_list[7]) + \
               (grades_list[8] * units_list[8])

# Instead of typing out "s1_u + s2_u + s3_u...",
# I used the built-in sum() function. It totals up my units list instantly.
total_units = sum(units_list)

# Normal division expression (/) to get my floating point GWA result
gwa = total_points / total_units


# SECTION 4: BUILT-IN LIST ANALYSIS (Chapters 3 & 8)

# Instead of writing a giant chain of 18 separate "or" conditions, I'm using max().
# Since a higher numerical score means a worse grade in my school system, 
# max(grades_list) automatically scans the whole envelope and isolates 
# my single worst grade.
worst_grade = max(grades_list)

# Now I only need to evaluate that one worst grade against my cutoff limits:
# Disqualified from DL/PL if my worst grade hits 2.50 or higher
dis_dl_pl = worst_grade >= 2.5

# Disqualified from Latin Honors if my worst grade hits 3.1 or higher
is_failing = worst_grade >= 3.1


# SECTION 5: OUTPUT DISPLAY

# Printing out the final academic evaluations.
print("ACADEMIC SUMMARY FOR:", name, " ")
print("Calculated GWA:", gwa)
print("Disqualified from DL/PL (Grade 2.5+):", dis_dl_pl)
print("Disqualified from Latin Honors (Grade 3.1+):", is_failing)
print("Congratss!!!")
print("You did your best! Padayon!")
```
